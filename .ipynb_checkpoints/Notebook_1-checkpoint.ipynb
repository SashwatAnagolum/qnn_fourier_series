#### Objective function

This objective function tries to make the model maximize the Hilbert-Schmidt distance between the two groups of statevectors that correspond to embedded data points from classes 0 and 1.

```python
def compute_objective(thetas, circ, data, var_params, embed_params):
    """
    Compute objective: compute the objective function value over the data using a linear loss function of the form 
    -f(x) * y where f(x) is the circuit output, and y is the class label.
    """
    losses = []
    class_0 = data[:len(data) // 2]
    class_1 = data[len(data) // 2:]
    class_0_vects = []
    class_1_vects = []
    
    for sample in class_0:
        qc = circ.assign_parameters({var_params[i]: thetas[i]
                                    for i in range(len(thetas))})
        qc = qc.assign_parameters({embed_params[i]: sample[i]
                                  for i in range(len(sample) - 1)})
        
        vect = Statevector.from_instruction(qc)
        class_0_vects.append(vect.data)

    class_0_vects = np.array(class_0_vects).reshape(len(class_0), len(vect.data))   
    class_0_mean = np.mean(class_0_vects, 0)
    class_0_dm = np.outer(np.conj(class_0_mean.T), class_0_mean)
        
    for sample in class_1:
        qc = circ.assign_parameters({var_params[i]: thetas[i]
                                    for i in range(len(thetas))})
        qc = qc.assign_parameters({embed_params[i]: sample[i]
                                  for i in range(len(sample) - 1)})
        
        vect = Statevector.from_instruction(qc)
        class_1_vects.append(vect.data)
        
    class_1_vects = np.array(class_1_vects).reshape(len(class_1), len(vect.data))   
    class_1_mean = np.mean(class_1_vects, 0)
    class_1_dm = np.outer(np.conj(class_1_mean.T), class_1_mean)
        
    hs_dist = np.trace((class_0_dm - class_1_dm) ** 2).real

    return 0.5 * (2 - hs_dist)
    

def compute_fidelity(qc, reg1, reg2, anc):
    """
    Compute fidelity: compute the fidelity of the states stored in the qubits indexed by reg1 and reg2 using the SWAP test,
    with the qubit indexed by anc as the ancilla.
    """
    if not qc.num_clbits:
        qc.add_register(ClassicalRegister(1))

    qc.h(anc)
    for index, qubit in enumerate(reg1):
        qc.cswap(anc, qubit, reg2[index])
        
    qc.h(anc)
    qc.measure(anc, 0)
    result = execute(qc, backend, shots=1024).result().get_counts()    
    fidelity = (result['0'] / 1024) - 0.5
    
    return fidelity
```

#### An alternative objective function

This objective function is from CLassification on Near-Term ... (Farhi, 2018) (add link).

```python
def compute_classification_objective(thetas, circ, data, var_params, embed_params, backend, verbose=0):
    """
    Compute classification objective: compute the MSE over the dataset used.
    """
    losses = []
    qc_var_fixed = circ.assign_parameters({var_params[i]: thetas[i] for i in range(len(thetas))})
    if not qc_var_fixed.num_clbits:
        qc_var_fixed.add_register(ClassicalRegister(1))
    
    for sample in data:
        qc = qc_var_fixed.assign_parameters({embed_params[i]: sample[i] for i in range(len(sample) - 1)})

        qc.measure(0, 0)
        result = execute(qc, backend, shots=1024).result().get_counts()
        label = int(sample[-1])
        
        if '1' in result.keys():
            prob = result['1'] / 1024
        else:
            prob = 0
            
        expectation_val = (2 * prob) - 1
        losses.append(-((2 * label) - 1) * expectation_val)
    
    if verbose == 1:
        print('MSE over data: {}'.format(sum(losses) / len(data)))
    elif verbose == 2:
        for i, sample in enumerate(data):
            print('Sample label: {} | Prediction: {}'.format((2 * sample[-1]) - 1, losses[i]))
        
    return sum(losses)
```

#### Minibatch gradient

Rather than calculate the gradient over the entire dataset, we can use a small batch of samples - this can help us escape local minima.

```python
def compute_minibatch_gradient(thetas, circ, data, var_params, embed_params, batch_size=5):
    """
    Compute minibatch gradient: compute the average gradient over batch_size samples, with equal samples
    from each class. If batch_size is odd, an extra sample is randomly chosen from a class.
    """
    if batch_size % 2:
        class_0_samples = (batch_size + 1) // 2 if np.random.sample() < 0.5 else batch_size // 2
    else:
        class_0_samples = batch_size // 2
        
    class_1_samples = batch_size - class_0_samples
    
    choices_0 = np.random.randint(0, len(data) // 2, class_0_samples)
    choices_1 = np.random.randint(len(data) // 2, len(data), class_1_samples)
    samples = data[np.concatenate((choices_0, choices_1))]
    
    gradients = []
    H = Z
    dim = len(samples[0]) - 1
    
    for sample in samples:
        qc = circ.assign_parameters({embed_params[i]: sample[i] for i in range(dim)})
        op = ~StateFn(H) @ CircuitStateFn(primitive=qc)
        gradient = Gradient(grad_method='param_shift').convert(operator=op, params=var_params).assign_parameters({
            var_params[i]: thetas[i] for i in range(len(thetas))
        }).eval()
        gradient = np.array(gradient)
        if sample[-1] == 0:
            gradient *= -1
            
        gradients.append(gradient.real)
       
    gradients = np.mean(np.array(gradients), 0)
    
    return gradients


def optimize(circ, init_point, steps, data, var_params, embed_params, backend, batch_size=1):
    """
    Optimize: optimize the circuit parameters with Nesterov momentum.
    """
    past_grads = [np.zeros(len(var_params))]
    objective_scores = []
    past_theta = np.copy(init_point)
    thetas = np.copy(init_point)
    lr = 0.005
    momentum = 0.9
    
    for i in range(steps):
#         prev_update = thetas - past_theta
#         past_thetas = np.copy(thetas)
#         thetas += momentum * prev_update
        grads = compute_minibatch_gradient(thetas, circ, data, var_params, embed_params, batch_size)
        thetas -= lr * grads
        
        print(compute_classification_objective(thetas, circ, data, var_params, embed_params, backend, 0))
        
    return thetas
    
```

#### Persistent Homology plotter

Because the default one doesn't look as nice.

```python
def plot_persistent_homology(data):
    """
    Plot persistent homoology: plot persistence barcodes for every n-dim 'hole' found.
    """
    max_0_dim = [max([sample[1] for sample in data[0][:-1]])]
    max_higher_dims = [max([sample[1] for sample in data[i]]) for i in range(1, len(data))]
    data[0][-1][1] = max(max_0_dim + max_higher_dims)
    plt.figure(figsize=[16, 9])
    plt.xlabel('Epsilon')
    plt.title('Persistent Homology')
    colors = []
    lines = []
    y = 0

    for dim in range(len(data)):
        color = np.random.sample(3)
        colors.append(color)
        lines.append(Line2D([0], [0], lw=4, color=color, label='Dimension: ' + str(dim)))
        for element in data[dim]:
            plt.plot([element[0], element[1]], [y, y], c=color)
            y += 1    

    plt.legend(handles=lines, loc='lower right')    
```

#### Create a variational circuit

Pass in the number of layers, qubits, and the kind of gates you want to use for embedding, rotation and entanglement blocks (this function only generates hardware efficient ansatze):

```python
def construct_feat_map(qc, dim, depth, use_second_order_embeddings):
    """
    Construct feature map: create an embedding circuit block with the gates specified, parametrized by the features of
    the data. If use_second_order_embeddings is True, then products of feature pairs are used to parametrize 
    the embedding block as well.
    """
    if use_second_order_embeddings:
        feat_map = ZZFeatureMap(dim, depth, 'linear', lambda x: np.pi * np.prod(x) * 0.5)
    else:
        feat_map = ZFeatureMap(dim, depth, lambda x: np.pi * x[0] * 0.5)
        
    embed_params = feat_map.parameters
    qc += feat_map
    
    return qc, list(embed_params)


def construct_circuit(num_qubits, depth, data_dim, rotation_gates=['ry', 'rz'], 
                      entanglement_gates=['cx'], use_second_order_embeddings=True, embedding_depth=1,
                      entanglement_map='linear'):
    """
    Construct circuit: construct a variational circuit based on the params specified.
    """
    qc = QuantumCircuit(num_qubits)
    qc, embed_params = construct_feat_map(qc, data_dim, embedding_depth, use_second_order_embeddings)
    var_params = []
    
    if depth:
        variational_block = TwoLocal(num_qubits, rotation_gates, entanglement_gates, entanglement_map, depth)
        var_params = list(variational_block.parameters)
        qc += variational_block

    return qc, var_params, embed_params
```

#### Postprocessing

These functions use the parameters learnt during the training process and capture the transformations affected by the circuit by looking at how the statevectors representing the embedded data points are transformed as they move through the circuit, and using these to create a distance matrix for computing persistent homology.

```python
def compute_intermediate_statevectors(num_layers, num_qubits, data, opt_params):
    """
    Compute intermediate statevectors: compute the statevectors affected by the circuit after every block of gates applied.    
    """
    high_dim_data = [[] for i in range(num_layers + 2)]
    
    for sample in data:
        qc = QuantumCircuit(num_qubits)
        for i in range(num_qubits):
            qc.ry(sample[i], i)

        for i in range(num_qubits - 1):
            qc.cx(i, i + 1)

        high_dim_data[0].append(Statevector.from_instruction(qc).data)

        for i in range(num_layers):
            for j in range(num_qubits):
                qc.ry(opt_params[(2 * i * num_qubits) + j], j)
                qc.rx(sample[j], j)
                qc.rz(opt_params[(((2 * i) + 1) * num_qubits) + j], j)

            for j in range(num_qubits - 1):
                qc.cx(j, j + 1)

            high_dim_data[i + 1].append(Statevector.from_instruction(qc).data)

        for j in range(num_qubits):
            qc.ry(opt_params[(2 * num_layers * num_qubits) + j], j)
            qc.rz(opt_params[(((2 * num_layers) + 1) * num_qubits) + j], j)

        high_dim_data[num_layers + 1].append(Statevector.from_instruction(qc).data)

    return np.array(high_dim_data)


def compute_hs_distances(vects):
    """
    Compute HS distances: compute the Hilbert-Schmidt distances between each vector passed in and return an 
    upper triangular distance matrix.
    """
    distances = []
    
    for i in range(vects.shape[0]):
        step_distances = np.zeros((vects.shape[1], vects.shape[1]))
        for j in range(vects.shape[1]):
            for k in range(j, vects.shape[1]):
                dm_1 = np.outer(np.conj(vects[i, j].T), vects[i, j])
                dm_2 = np.outer(np.conj(vects[i, k].T), vects[i, k])
                hs_dist = np.trace((dm_1 - dm_2) ** 2).real
                step_distances[j, k] = hs_dist
        
        distances.append(step_distances)
        
    return distances
```

#### Functions to visualize the data transformation

These functions helps us visualize the transformations affected by the circuit by plotting persistence diagrams and PCA-transformed low dimensional representations of the data.

```python
def plot_ph_changes(dms, max_dim=1):
    """
    Plot persistent homology changes: plot the persistance diagrams based on the transformed statevectors representing the
    embedded data after every block.
    """
    for i in range(len(dms)):
        ph_result = ripser(dms[i], distance_matrix=True, maxdim=max_dim)['dgms']
        plot_persistent_homology(ph_result)
        
        
def plot_pca_transformed_data(vects, num_samples, use_3d=True):
    """
    Plot low dimensional representations of the embedded data points.
    Warning: Currently broken - plots real and imaginary parts of statevectors separately.
    """
    dim = 3 if use_3d else 2
    pca = PCA(n_components=dim)

    for i in range(len(vects)):
        real_points = pca.fit_transform(vects[i].real)
        imag_points = pca.fit_transform(vects[i].imag + 1e-8)
        fig = plt.figure(figsize=[16, 9])
        
        if use_3d:
            ax = fig.add_subplot(1, 2, 1, projection='3d')
            ax2 = fig.add_subplot(1, 2, 2, projection='3d')
            ax.scatter(real_points[:num_samples, 0], real_points[:num_samples, 1], real_points[:num_samples, 2], c='b')
            ax.scatter(real_points[num_samples:, 0], real_points[num_samples:, 1], real_points[num_samples:, 2], c='r')    
            ax2.scatter(imag_points[:num_samples, 0], imag_points[:num_samples, 1], imag_points[:num_samples, 2], c='b')
            ax2.scatter(imag_points[num_samples:, 0], imag_points[num_samples:, 1], imag_points[num_samples:, 2], c='r') 
        else:
            ax = fig.add_subplot(1, 2, 1)
            ax2 = fig.add_subplot(1, 2, 2)
            ax.scatter(real_points[:num_samples, 0], real_points[:num_samples, 1], c='b')
            ax.scatter(real_points[num_samples:, 0], real_points[num_samples:, 1], c='r')    
            ax2.scatter(imag_points[:num_samples, 0], imag_points[:num_samples, 1], c='b')
            ax2.scatter(imag_points[num_samples:, 0], imag_points[num_samples:, 1], c='r') 

        plt.show()
```

Why do we use Hilbert-Schmidt:





```python
def fubini_study_metric(state1, state2):
    return np.arccos(np.abs(np.dot(np.conj(state1.T), state2)))

def hilbert_schmidt_metric(state1, state2):
    return np.trace((np.outer(np.conj(state1.T), state1) - np.outer(np.conj(state2.T), state2)) ** 2).real

def trace_dist(state1, state2):
    return 0.5 * np.trace(np.abs(np.outer(np.conj(state1.T), state1) - np.outer(np.conj(state2.T), state2))).real

def euclidean_dist(state1, state2):
    return np.sqrt(sum([np.conj(state1[i] - state2[i] * (state1[i] - state2[i])) for i in range(len(state1))]))
```
